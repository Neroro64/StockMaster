```python
 pca = PCA(n_components=5)
 period = data[1600:2600]

 X = period[["inter-diff","inter-derivative", "1-rad", "adx", "macd", "macdsignal", "rsi", "willr"]]
 X["BB"] = period["close"] - period["BB-middle"]
 X["ema"] = period["ema5"] - period["ema20"]
 X["sma"] = period["sma50"] - period["sma200"]
 X["sar"] = period["close"] - period["sar"]
 X["stoch"] = period["stoch_k"] - period["stoch_d"]

pca.fit(X.T)
print(pca.components_)
print(pca.explained_variance_ratio_)
print(pca.get_covariance())
```
